```python
# labelling and converting the images into numpy arrays
import numpy as np
import glob

cat_path='/content/kagglecatsanddogs_3367a/New_Cat'
dog_path='/content/kagglecatsanddogs_3367a/New_Dog'

All_Images=[]
Labels=[]

for filename in glob.glob(os.path.join(cat_path,'*.*')):
  img=cv2.imread(filename)
  All_Images.append(img)
  Labels.append(0)

for filename in glob.glob(os.path.join(dog_path,'*.*')):
  img=cv2.imread(filename)
  All_Images.append(img)
  Labels.append(1)

# convert thelist of all images and labels into numpy arrays
All_Images=np.asarray(All_Images)
Labels=np.asarray(Labels)

# shuffle the data
from sklearn.utils import shuffle
All_Images,Labels=shuffle(All_Images,Labels)

# verify the shuffling
print('first 10 shuffled labels',Labels[:10])
print('first 10 shuffled images',All_Images[:10])
```

```python
# dividing the features and the target

x=All_Images
y=Labels
```

```python
# Splitting the data into training and testing data

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from keras.utils import to_categorical
y_train=to_categorical(y_train)
y_test=to_categorical(y_test)
```

```python
# Scalling the features

x_train=x_train/255
x_test=x_test/255
```

```python
# Building a convolutional neural network

from keras.models import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten

model=Sequential()

model.add(Conv2D(32,kernel_size=(3,3),activation='relu',strides=(1,1),input_shape=(245,245,3)))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))

model.add(Conv2D(64,kernel_size=(3,3),activation='relu',strides=(1,1)))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))

model.add(Conv2D(128,kernel_size=(3,3),activation='relu',strides=(1,1)))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))

model.add(Flatten())

model.add(Dense(300,activation='relu'))
model.add(Dense(2,activation='softmax'))

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

model.fit(x_train,y_train,epochs=5)
```

```python
# testing accuracy
loss,accuracy=model.evaluate(x_test,y_test)
print('loss:',loss)
print('accuracy:',accuracy)
```

```python
# making a preictive system
input_path=input('Enter the path of the image:')
img=plt.imread(input_path)
img=cv2.resize(img,(245,245))
if img.ndim == 2:  # If grayscale, convert to BGR
    img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
plt.imshow(img)
plt.show()
img=img/255
img=np.reshape(img,(1,245,245,3))
result=model.predict(img)
result=np.argmax(result)
if result==0:
  print('Cat')
else:
  print('Dog')
```
