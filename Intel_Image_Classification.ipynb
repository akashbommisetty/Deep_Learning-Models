```python
import glob
buildings="/content/intel-image-classification/seg_train/seg_train/buildings"
glacier="/content/intel-image-classification/seg_train/seg_train/glacier"
forest="/content/intel-image-classification/seg_train/seg_train/forest"
sea="/content/intel-image-classification/seg_train/seg_train/sea"
mountain="/content/intel-image-classification/seg_train/seg_train/mountain"
street="/content/intel-image-classification/seg_train/seg_train/street"

All_Images=[]
Labels=[]

Buildings_count=0
for filename in glob.glob(os.path.join(buildings, '*.*')):
  if Buildings_count>200:
    break
  img=cv2.imread(filename)
  img=cv2.resize(img,(150,150))
  All_Images.append(img)
  Labels.append(0)
  Buildings_count+=1

Glacier_count=0
for filename in glob.glob(os.path.join(glacier, '*.*')):
  if Glacier_count>200:
    break
  img=cv2.imread(filename)
  img=cv2.resize(img,(150,150))
  All_Images.append(img)
  Labels.append(1)
  Glacier_count+=1

Forest_count=0
for filename in glob.glob(os.path.join(forest, '*.*')):
  if Forest_count>200:
    break
  img=cv2.imread(filename)
  img=cv2.resize(img,(150,150))
  All_Images.append(img)
  Labels.append(2)
  Forest_count+=1

Sea_count=0
for filename in glob.glob(os.path.join(sea, '*.*')):
  if Sea_count>200:
    break
  img=cv2.imread(filename)
  img=cv2.resize(img,(150,150))
  All_Images.append(img)
  Labels.append(3)
  Sea_count+=1

Mountain_count=0
for filename in glob.glob(os.path.join(mountain, '*.*')):
  if Mountain_count>200:
    break
  img=cv2.imread(filename)
  img=cv2.resize(img,(150,150))
  All_Images.append(img)
  Labels.append(4)
  Mountain_count+=1

Street_count=0
for filename in glob.glob(os.path.join(street, '*.*')):
  if Street_count>200:
    break
  img=cv2.imread(filename)
  img=cv2.resize(img,(150,150))
  All_Images.append(img)
  Labels.append(5)
  Street_count+=1

from sklearn.utils import shuffle
All_Images,Labels=shuffle(All_Images,Labels)
```

```python
x=All_Images
y=Labels
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
import numpy as np # Import NumPy

x_train = np.array(x_train) # Convert x_train to NumPy array
x_test = np.array(x_test) # Convert x_test to NumPy array

x_train = x_train / 255.0  # Perform element-wise division
x_test = x_test / 255.0  # Perform element-wise division
```

```python
from keras.utils import to_categorical
y_train = to_categorical(y_train, num_classes=6)
y_test=to_categorical(y_test)
```

```python
x_train.shape
```

```python
import keras
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Flatten,Dense,Dropout
from sklearn.metrics import accuracy_score

model=Sequential()
model.add(Conv2D(64,(3,3),activation="relu",input_shape=(150,150,3)))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(128,(3,3),activation="relu"))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(128,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(6,activation="softmax"))
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,batch_size=10,epochs=10)
```

```python
#Making a predictive system
input_path=input('Enter the path of the image:')
img=plt.imread(input_path)
img=cv2.resize(img,(150,150))
if img.ndim == 2:  # If grayscale, convert to BGR
    img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
plt.imshow(img)
plt.show()
# img=preprocess_input(img)
#The line below reshapes the image to have an additional dimension representing the batch size.
img=np.reshape(img,(1,150,150,3)) # Reshape to (1, 150, 150, 3)
result=model.predict(img)
result=np.argmax(result)
if result==0:
    print('Building')
elif result==1:
    print('glacier')
elif result==2:
    print('forest')
elif result==3:
    print('sea')
elif result==4:
    print('mountain')
else:
  print("street")
```
