```python
import glob
import cv2
import os
import numpy as np
import pandas as pd

# Paths to the directories
dogs_path = 'kaggle-cat-vs-dog-dataset/kagglecatsanddogs_3367a/PetImages/Dog'
cats_path = 'kaggle-cat-vs-dog-dataset/kagglecatsanddogs_3367a/PetImages/Cat'

All_Images = []
Labels = []

# Load and process dog images
cats_count = 0
for filename in glob.glob(os.path.join(cats_path, '*.*')):
    if cats_count >= 3000:
        break
    img = cv2.imread(filename)
    if img is None:
        print(f"Warning: {filename} could not be loaded.")
        continue
    img = cv2.resize(img, (224, 224))
    All_Images.append(img)
    Labels.append(0)
    cats_count += 1

# Load and process cat images
dogs_count = 0
for filename in glob.glob(os.path.join(dogs_path, '*.*')):
    if dogs_count >= 3000:
        break
    img = cv2.imread(filename)
    if img is None:
        print(f"Warning: {filename} could not be loaded.")
        continue
    img = cv2.resize(img, (224, 224))
    All_Images.append(img)
    Labels.append(1)
    dogs_count += 1

# Convert lists to numpy arrays
All_Images = np.asarray(All_Images)
Labels = np.asarray(Labels)

print("Number of images:", len(All_Images))
print("Number of labels:", len(Labels))

from sklearn.utils import shuffle
All_Images,Labels=shuffle(All_Images,Labels)
print('first 10 shuffled labels',Labels[:10])
print('first 10 shuffled images',All_Images[:10])
```

```python
# dividing the features and the target

x=All_Images
y=Labels
```

```python
from keras.applications import VGG19
from keras.applications.vgg19 import preprocess_input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
x = preprocess_input(x)
```

```python
# Splitting the data into training and testing data

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from keras.utils import to_categorical
y_train=to_categorical(y_train)
y_test=to_categorical(y_test)
```

```python
# Data augmentation
train_datagen = ImageDataGenerator(rotation_range=20, zoom_range=0.15, width_shift_range=0.2,
                                   height_shift_range=0.2, shear_range=0.15, horizontal_flip=True,
                                   fill_mode="nearest")
```

```python
# Building the VGG19-based model
base_model = VGG19(weights='imagenet', include_top=False, input_shape=(224,224, 3))

# Freezing the layers of VGG19 model
for layer in base_model.layers:
    layer.trainable = False

# Adding custom layers on top of VGG19
model = Sequential()
model.add(base_model)
model.add(Flatten())
model.add(Dense(300, activation='relu'))
model.add(Dense(2, activation='softmax'))  # 2 classes: Cat and Dog

# Compiling the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Training the model
model.fit(train_datagen.flow(x_train, y_train, batch_size=32), epochs=2, validation_data=(x_test, y_test))
```
